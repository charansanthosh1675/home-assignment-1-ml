```python
import numpy as np
import matplotlib.pyplot as plt

# Generate dataset
np.random.seed(0)
x = np.random.uniform(0, 5, 200)
noise = np.random.normal(0, 1, 200)
y = 3 + 4*x + noise

# Add bias term
X = np.c_[np.ones(len(x)), x]

# Closed-form solution (Normal Equation)
theta_closed = np.linalg.inv(X.T @ X) @ X.T @ y

# Gradient Descent
theta = np.zeros(2)
eta = 0.05
iterations = 1000
losses = []

for _ in range(iterations):
    y_pred = X @ theta
    errors = y_pred - y
    gradient = (2/len(y)) * X.T @ errors
    theta = theta - eta * gradient
    losses.append(np.mean(errors**2))

# Plot raw data and fitted lines
plt.scatter(x, y, label="Raw Data")
plt.plot(x, X @ theta_closed, label="Closed-form Line")
plt.plot(x, X @ theta, label="Gradient Descent Line")
plt.legend()
plt.show()

# Loss curve
plt.plot(losses)
plt.xlabel("Iterations")
plt.ylabel("MSE Loss")
plt.title("Gradient Descent Loss Curve")
plt.show()

print("Closed-form theta:", theta_closed)
print("Gradient Descent theta:", theta)
```
